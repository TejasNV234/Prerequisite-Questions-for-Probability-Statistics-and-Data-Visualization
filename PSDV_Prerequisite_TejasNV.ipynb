```python
# 1. List Operations and Basic Statistics
from math import sqrt
nums = [x for x in range(1, 21)]
sumEvenNums = 0
prodOddNums = 1
for x in nums:
    if x%2==0:
        sumEvenNums += x
    else:
        prodOddNums *= x
meanNums = sum(nums)/len(nums)
medianNums = (nums[len(nums)//2 - 1] + nums[len(nums)//2])/2
standardDeviation = sqrt(sum([(x-meanNums)**2 for x in nums])/len(nums))

print(f"""
Sum of Even Numbers: {sumEvenNums}
Product of Odd Numbers: {prodOddNums}
Mean: {meanNums}
Median: {medianNums}
Standard Deviation: {standardDeviation}
""")
```

```python
# 2. Mutable vs. Immutable Data Types

# Objects of Mutable Datatypes can be changed once created
# Objects of Immutable Datatypes can't be changed once created

# Mutable Datatype
lst = [1, 2]
print(lst)
lst[1] = 90
print(lst)

# Immutable Datatype
name = "Tejas NV"
try:
    name[5] = "_"
except:
    print("Couldn't change string, therefore Immutable Datatype")
```

```python
# 3. Fibonacci Sequence
import matplotlib.pyplot as plt

num = int(input())

def fibonacci(n):
    series = []
    a, b = 0, 1
    while(len(series) < n):
        series.append(a)
        a, b = b, a+b
    return series

print(fibonacci(num))

plt.plot(fibonacci(num), linestyle="dotted")
plt.grid(True)
plt.show()
```

```python
# 4. List Manipulation

lst = [3, 5, 7, 9, 11, 13]
print(f"Given List:    {lst}")

lst.insert(2, 6)
print(f"Modified List: {lst}")

lst.remove(7)
print(f"Modified List: {lst}")

lst = lst[::-1]
print(f"Modified List: {lst}")

meanLst = sum(lst)/len(lst)
medianLst = (lst[len(lst)//2 - 1] + lst[len(lst)//2])/2
print(f"""
Mean of List: {meanLst}
Median of List: {medianLst}
""")
```

```python
# 5. User Input and Basic Statistics
import matplotlib.pyplot as plt

inputStr = input()
lst = [int(x) for x in inputStr.split(", ")]

meanLst = sum(lst)/len(lst)

minLst = lst[1]
for x in lst:
    if minLst > x:
        minLst = x

maxLst = lst[1]
for x in lst:
    if maxLst < x:
        maxLst = x

print(f"""
Mean of List: {meanLst}
Minimum of List: {minLst}
Maximum of List: {maxLst}
""")

plt.hist(lst, color='skyblue', edgecolor="black", align='mid', rwidth=0.5)
plt.xlabel("Range of Values")
plt.ylabel("Frequency")
plt.title("Histogram of Values in List")

plt.show()
```

```python
# 9: Merging Dictionaries

dict1 = {'a': 10, 'b': 20}
dict2 = {'b': 5, 'c': 15}
dict3= {}

for x, y in (dict1 | dict2).items():
    if x in dict1 and x in dict2:
        dict3[x] = y+dict1.get(x)
    else:
        dict3[x] = y

print(dict3)
```

```python
# 10: Character Frequency

import matplotlib.pyplot as plt
def func():
    freq = {}
    inputStr = input()
    inputStr = inputStr.lower()

    for x in inputStr:
        if x not in freq:
            freq[x] = inputStr.count(x)
    
    plt.bar(freq.keys(), freq.values())
    plt.xlabel("Characters (blank is the space character)")
    plt.ylabel("Frequency")
    plt.show()

    return freq

func()
```

```python
# 11: Remove Duplicates

inputStr1 = input()

lst1 = [int(x) for x in inputStr1.split()]
lst3 = []

for x in lst1:
    if x not in lst3:
        lst3.append(x)

medianUniqueElem = 0

if len(lst3) % 2 == 0:
    medianUniqueElem = (lst3[len(lst3)//2 - 1] + lst3[len(lst3)//2])/2
else:
    medianUniqueElem = lst3[(len(lst3) + 1)//2-1]

print(f"""
Mean of Unique Elements: {sum(lst3)/len(lst3)}
Median of Unique Elements: {medianUniqueElem}
""")
```

```python
# 12: String Operations

import matplotlib.pyplot as plt

stringLst = ['jkjk', 'jkjk', 'awow', 'uwow', 'uiuiusidi', 'fdsjfhjksa', 'dasfjiubj']
stringLengths = {}

lenLongest = 0
longest = ''
countVowelStart = 0
for x in stringLst:
    if len(x) > lenLongest:
        lenLongest = len(x)
        longest = x
    if x[0] in ['a', 'e', 'i', 'o', 'u']:
        countVowelStart += 1
    stringLengths[x] = len(x)

print(stringLengths, end="\n\n")
print(f"Longest String is: {longest}")
print(f"Number of strings starting with vowel: {countVowelStart}\n")

plt.hist(stringLengths.values(), color='skyblue', edgecolor='black')
plt.xlabel("String Lengths")
plt.ylabel("Frequency")
plt.title("Frequency of String Lengths")
plt.show()
```

```python
# 15: Anagram Finder

word = input("Word: ")
words = input('Words separated by ", ": ')

words = words.split(", ")
def anagramFinder(word, words):
    word = sorted(word)
    anagrams = []
    for x in words:
        if word == sorted(x):
            anagrams.append(x)
    return anagrams

print(f"\nAnagrams of the word '{word}' in the words given are: {anagramFinder(word, words)}")
```

```python
# 16: Set Operations
from matplotlib_venn import venn2,venn2_circles 

lst1 = [x for x in range(1, 21, 2)]
lst2 = [x for x in range(2, 21)]

union = list(set(lst1+lst2))
intersection = []
symmetricDiff = []

for x in union:
    if x in lst1 and x in lst2:
        intersection.append(x)
for x in union:
    if x not in intersection:
        symmetricDiff.append(x)
print(f"""
Union: {union}
Intersection: {intersection}
Symmetric Difference: {symmetricDiff}
 """)

venn2(subsets = (set(lst1), set(lst2)), set_labels = ('lst1', 'lst2'),set_colors=("yellow", 
                  "red"),alpha=0.7) 
venn2_circles(subsets=(set(lst1), set(lst2)))
plt.show()
```

```python
# 19: Random Number Generation
from random import randint
import matplotlib.pyplot as plt

randInts = []

while(len(randInts) < 100):
    randInts.append(randint(1, 50))

with open("RandomIntegers.txt", 'w') as f:
    f.write("\n".join([str(x) for x in randInts]))

with open("RandomIntegers.txt", 'r') as f:
    data = [int(x) for x  in f.read().split("\n")]
    mean = sum(data)/len(data)

    print(f"Mean of the Numbers: {mean}")
    
    plt.figure().set_figwidth(20)
    plt.bar(data, [data.count(x) for x in data])
    plt.show()
```

```python
# 21: Basic Statistics
from random import randint
from math import sqrt
import matplotlib.pyplot as plt

lstOfNums = [randint(0, 150) for x in range(150)]

mean = sum(lstOfNums)/len(lstOfNums)
median = 0

if len(lstOfNums) %2 == 0:
    median = sorted(lstOfNums)[len(lstOfNums)//2 - 1] + sorted(lstOfNums)[len(lstOfNums)//2]
else:
    median = sorted(lstOfNums)[(len(lstOfNums) - 1)//2]

sd = sqrt(sum([((x - mean) ** 2) for x in lstOfNums]) / len(lstOfNums))

print(f"""
The list: {lstOfNums}
Mean: {mean}
Median: {median}
Standard Deviation: {sd}
""")

plt.boxplot(lstOfNums)
plt.show()
```

```python
# 22: Normalization
from random import randint
import matplotlib.pyplot as plt

k = randint(0, 20)
lstNums = sorted([x for x in range(k, k+21)])
print(f"Original List: {lstNums}\n")

lstNums = [round(x/max(lstNums), 2) for x in lstNums]
print(f"Normalised List: {lstNums}\n")

plt.plot(lstNums, linestyle="dotted")
plt.xticks([x for x in range(1, 21)])
plt.yticks([x/10 for x in range(1, 11)])
plt.show()
```
